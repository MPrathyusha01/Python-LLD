# Variabe

Variables are containers for storing data values.

Python has no command for declaring a variable. A variable is created the moment you first assign a value to it.

* *Rules for Variable Names:*
1. Allowed Characters:
Variable names can contain letters (a-z, A-Z), numbers (0-9), and underscores (_). 
2. First Character:
The first character of a variable name must be a letter (a-z, A-Z) or an underscore (_). 
3. No Spaces:
Variable names cannot contain spaces. 
4. Case Sensitivity:
Python is case-sensitive, meaning my_variable and My_Variable are considered different variables. 
5. Keywords:
You cannot use Python's reserved keywords (like if, else, for, while, def, etc.) as variable names. 
6. Descriptive Names:
Choose descriptive names that reflect the data the variable represents (e.g., student_count instead of sc). 
7. Snake Case:
Use snake\_case (e.g., my_variable_name) for variable names, where words are separated by underscores.

# Casting

* int() - constructs an integer number from an integer literal, a float literal (by removing all decimals), or a string literal (providing the string represents a whole number)
* float() - constructs a float number from an integer literal, a float literal or a string literal (providing the string represents a float or an integer)
* str() - constructs a string from a wide variety of data types, including strings, integer literals and float literals
  
Note: You cannot convert complex numbers into another number type.

```python
x = int(1)   # x will be 1
y = int(2.8) # y will be 2
z = int("3") # z will be 3

print(x,y,z)
print(type(x))
print(type(y))
print(type(z))
```

## Python String Methods Cheat Sheet

| **Method**              | **Definition**                                          | **Example**                                   |
|-------------------------|----------------------------------------------------------|-----------------------------------------------|
| `capitalize()`          | Capitalizes the first letter of the string.              | `"hello".capitalize()  'Hello'`            |
| `casefold()`            | Converts string to lowercase (stronger than lower).      | `"Hello".casefold()  'hello'`              |
| `center(width)`         | Centers the string in given width with spaces.           | `'py'.center(5)  '  py '`                  |
| `count(sub)`            | Counts occurrences of a substring.                       | `"banana".count('a')  3`                   |
| `endswith(suffix)`      | Returns True if string ends with suffix.                 | `"data.csv".endswith('.csv')  True`        |
| `find(sub)`             | Finds index of first occurrence, -1 if not found.        | `"apple".find('p')  1`                     |
| `format()`              | Formats strings using placeholders.                      | `"My name is {}".format('Bestie')`           |
| `index(sub)`            | Like find() but raises error if not found.               | `"apple".index('p')  1`                    |
| `isalnum()`             | True if string has only alphanumeric characters.         | `"abc123".isalnum()  True`                 |
| `isalpha()`             | True if string has only alphabets.                       | `"abc".isalpha()  True`                    |
| `isdigit()`             | True if string has only digits.                          | `"1234".isdigit()  True`                   |
| `islower()`             | True if all characters are lowercase.                    | `"hello".islower()  True`                  |
| `isupper()`             | True if all characters are uppercase.                    | `"HELLO".isupper()  True`                  |
| `isspace()`             | True if string contains only spaces.                     | `"   ".isspace()  True`                    |
| `istitle()`             | True if string is in title case.                         | `"Hello World".istitle()  True`            |
| `join(iterable)`        | Joins elements of iterable with string as separator.     | `"-".join(['a','b'])  'a-b'`               |
| `lower()`               | Converts entire string to lowercase.                     | `"Hello".lower()  'hello'`                 |
| `lstrip()`              | Removes leading spaces.                                 | `"   hi".lstrip()  'hi'`                   |
| `rstrip()`              | Removes trailing spaces.                                | `"hi   ".rstrip()  'hi'`                   |
| `strip()`               | Removes leading and trailing spaces.                    | `"  hello  ".strip()  'hello'`             |
| `replace(old, new)`     | Replaces old substring with new.                         | `"hello".replace('l','p')  'heppo'`        |
| `rfind(sub)`            | Finds last occurrence of substring.                      | `"hello hello".rfind('l')  9`              |
| `split(sep)`            | Splits string by separator into list.                    | `"a,b,c".split(',')  ['a','b','c']`        |
| `rsplit(sep)`           | Splits from right by separator.                          | `"a,b,c".rsplit(',',1)  ['a,b','c']`       |
| `startswith(prefix)`    | True if string starts with prefix.                       | `"Hello".startswith('H')  True`            |
| `swapcase()`            | Swaps lower to upper and vice versa.                     | `"HeLLo".swapcase()  'hEllO'`              |
| `title()`               | Converts string to title case.                           | `"hello world".title()  'Hello World'`     |
| `upper()`               | Converts entire string to uppercase.                     | `"hello".upper()  'HELLO'`                 |
| `zfill(width)`          | Pads string with zeros on the left to reach width.       | `"42".zfill(5)  '00042'`                   |
